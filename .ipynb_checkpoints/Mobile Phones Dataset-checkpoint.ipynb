# MOBILE DATASET 2025

```python
# Import the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('Mobiles Dataset (2025).csv', encoding='latin1')
df.head()
```

```python
# Get the summary of the dataset
df.info()
```

```python
# Change the value type of some columns from string to float
numeric_columns = ['Mobile Weight', 'RAM', 'Front Camera', 'Back Camera', 'Battery Capacity', 
                   'Launched Price (Pakistan)', 'Launched Price (India)', 'Launched Price (China)', 'Launched Price (USA)', 
                   'Launched Price (Dubai)']

for col in numeric_columns:
    df[col] = df[col].str.replace(r'[^\d]', '', regex=True)
    df[col] = df[col].replace('', 'NaN')
    df[col] = df[col].astype(float)
```

```python
expressions = ['inches','(main)','(internal)','(external)','(unfolded)', '()']
for x in expressions:
    df['Screen Size'] = df['Screen Size'].str.replace(x, '', regex=True)

df['Screen Size'] = df['Screen Size'].apply(lambda x: x.split(' ')[0]).astype(float)
```

```python
# Add the units of measurement into the relevant columns
df.rename(columns={'Mobile Weight':'Mobile Weight (grams)', 
                   'Front Camera':'Front Camera (MP)',
                   'RAM':'RAM (GB)',
                   'Back Camera': 'BackCamera (MP)',
                  'Battery Capacity':'Battery Capacity (mAh)',
                  'Screen Size':'Screen Size (inches)'}, inplace=True)
```

```python
df.head()
```

```python
# Plot chart for the RAM Distribution
sns.set_style("whitegrid")
sns.set_palette("coolwarm")
plt.figure(figsize=(15, 5))
sns.countplot(x='RAM (GB)', data=df, edgecolor='black')
plt.xlabel('RAM (GB)')
plt.ylabel('Distribution')
plt.title('RAM Distribution')
plt.xticks(rotation=0);
plt.show()
```

```python
# Plot the market share of the mobile companies
plt.figure(figsize=(15,5))
sns.barplot(x='Company Name', y='Launched Price (USA)', data=df, edgecolor='black')

plt.xticks(rotation=45)
plt.xlabel('Mobile Companies')
plt.ylabel('Market Share')
plt.title('Mobile Companies Market Share')

plt.show()
```

```python
# Plot the number of phone brands per Company
plt.figure(figsize=(15, 5))

sns.countplot(x='Company Name', data=df, order=df['Company Name'].value_counts().index, edgecolor='black')

plt.xticks(rotation=45)
plt.xlabel('Mobile Companies')
plt.ylabel('Frequency')
plt.title('Distribution of Mobile Companies')

plt.show()
```

```python
top_5_phones = df['Company Name'].value_counts().head().index.tolist()
top_5 = df['Company Name'].value_counts().head(5).sort_values(ascending=False)
top_5_lst = top_5.index.tolist()
top_5_lst.insert(0, 'None')
```

```python
def expensive_phones(Company):
    company_models = df[df['Company Name'] == Company][['Model Name', 'Launched Price (USA)']]
    return company_models.sort_values(by='Launched Price (USA)', ascending=False)

oppo = expensive_phones('Oppo')
apple = expensive_phones('Apple')
honor = expensive_phones('Honor')
samsung = expensive_phones('Samsung')
vivo = expensive_phones('Vivo')

popular = ['None', oppo, apple, honor, samsung, vivo]
```

```python
# Plot the most popular Phone Companies together with their most expensive phone brands
fig, ax = plt.subplots(figsize=(20, 20), nrows=3, ncols=2)

sns.set_style('ticks')
sns.barplot(x=top_5.values, y=top_5.index, ax=ax[0, 0], orient='h')
ax[0, 0].set_title('Top 5 most manufactured Phone brands')

for i in range(1, 6):
    row = i // 2
    col = i % 2
    models = popular[i].head(10)
    sns.barplot(x='Launched Price (USA)', y='Model Name', data=models, ax=ax[row, col], orient='h',edgecolor='black')
    ax[row, col].set_title(f'Top ten most expensive {top_5_lst[i]} phones')

plt.show()
```

```python
# Overall most expensive phone brands
expensive_phones = df.nlargest(10, 'Launched Price (USA)')[['Model Name', 'Launched Price (USA)']]
expensive_phones
```

```python
# Cheapest phone brands
cheapest_phones = df.nsmallest(10, 'Launched Price (USA)')[['Model Name', 'Launched Price (USA)']]
cheapest_phones
```

```python
# Distribution of mobile phone RAM
df = df[(df['Screen Size (inches)'] <=10) & (df['Screen Size (inches)'] >= 2)]

plt.figure(figsize=(15, 5))

sns.histplot(df['Screen Size (inches)'], bins=15, kde=False, edgecolor='black')

xticks = np.linspace(5, 12, 8)
plt.xticks(xticks)
plt.xlim(4, 13)
plt.xlabel('Screen Size (inches)')
plt.ylabel('Frequency')
plt.title('Screen Size Distribution')

plt.show()
```

```python
# Ten most common processors
ten_most_common_processors = df['Processor'].value_counts().head(10)

plt.figure(figsize=(15, 7))
sns.barplot(x=ten_most_common_processors.values, y=ten_most_common_processors.index, orient='h', edgecolor='black')

plt.xlabel('Frequency')
plt.ylabel('Processors')
plt.title('Distribution of Mobile Phone Processors')

plt.show()
```

```python
# Converting the different currencies to a common currency (USD)
prices = df[['Model Name','Launched Price (USA)', 'Launched Price (Pakistan)', 
             'Launched Price (India)','Launched Price (China)','Launched Price (Dubai)']]

prices.loc[:, 'Launched Price (Pakistan)'] = prices['Launched Price (Pakistan)'] * 0.003570
prices.loc[:, 'Launched Price (India)'] = prices['Launched Price (India)'] * 0.0115955
prices.loc[:, 'Launched Price (China)'] = prices['Launched Price (China)'] * 0.1418
prices.loc[:, 'Launched Price (Dubai)'] = prices['Launched Price (Dubai)'] * 0.272
```

```python
# Launch prices in diffrerent countries in a common currency (USD)
plt.figure(figsize=(8,4))
sns.boxplot(data=prices[['Launched Price (USA)', 'Launched Price (Pakistan)', 'Launched Price (India)',
                      'Launched Price (China)','Launched Price (Dubai)']], showfliers=False, palette='muted')
labels=['USA', 'Pakistan', 'India', 'China', 'Dubai']
plt.xticks(ticks=[0,1,2,3,4], labels=labels)
plt.grid(True);
```

```python
# Battery Size trend over time
sns.lineplot(x='Launched Year', y='Battery Capacity (mAh)', data=df)
plt.title('Battery Capacity Trend Over Time')
plt.grid(True);
```
